### Example 1: Lets say we want to delete each element in the list which is divisible by 2 or all the even numbers.

```python
list1 = [11, 5, 17, 18, 23, 50] 
for i in list1:
    if (i % 2 == 0):
        list1.remove(i)
list1
```

```python
#Using list comp
list1 = [11, 5, 17, 18, 23, 50] 
ele_lst = [i for i in list1 if  (i % 2 != 0)]
print(ele_lst)
```

### Example 2: Remove adjacent elements using list slicing. Below Python code remove values from index 1 to 4. 

```python
list1 = [11, 5, 17, 18, 23, 50] 
del list1[1:5]
list1
```
