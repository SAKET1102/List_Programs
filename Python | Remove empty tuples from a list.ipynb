```python
tuples = [(), ('ram','15','8'), (), ('laxman', 'sita'), 
          ('krishna', 'akbar', '45'), ('',''),()]
tuples
```

```python
#Using filter method
tuples = list(filter(None,tuples))
tuples
```

```python
#Using for loop
for t in tuples:
    if t:
        print(t)
```
