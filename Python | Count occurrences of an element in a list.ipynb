```python
#Using for Loop
lst = [15, 6, 7, 10, 12, 20, 10, 28, 10]
num = 10
cnt = 0
for i in lst:
    if i == num:
        cnt = cnt + 1
print(f"number of occurrences of {num} in list{lst} is:{cnt}")
```

```python
lst = [15, 6, 7, 10, 12, 20, 10, 28, 10]
num = 10
from collections import Counter
Count = Counter(lst)
print(f"number of occurrences of {num} in list{lst} is:{Count[6]}")
```
